```python
def input_invalid_int(msg, start = 0, end = None):
#     keep iterating till the given input is valid
    print(msg)
    while True:
        inp=input()
        
        if not inp.isdecimal():
            print('Invalid input. please try again ^_^ !')
        elif start is not None and end is not None:
            if not(start<= int(inp) <= end):
                print('Invalid input range. please try again ^_^ !')
            else:
                return int(inp)
#         if correct input     
        else:
            return int(inp)
    
```

```python
class Employee:
    def __init__(self, name, age, salary):
        self.name, self.age, self.salary = name, age, salary
    
    def __str__(self):
        return f'Employee(name="{self.name}", age={self.age}, salary={self.salary})'
    
    def __repr__(self):
        return F'Employee(name="{self.name}", age={self.age}, salary={self.salary})'
```

```python
class EmployeeManager:
    def __init__(self):
        self.employees = []
        
    def add_employee(self):
        print("\nEnter Employee Data: ")
        name = input('Enter The Name: ')

#         print("Enter The Age: ")
        age = input_invalid_int('Enter The Age: ')
        
#         print("Enter Salary")
        salary =  input_invalid_int('Enter The Salary: ')

        self.employees.append(Employee(name, age, salary))
        
#         List of all Employees
    def list_employees(self):
#         Check if we have employees or not 
        if len(self.employees)==0:
            print('\nNo Employees at the moment.')
            return
#         otherwise list of employees
        print('\n**Employees list**')
        for emp in self.employees:
            print(emp)
    
#     delete employee
    def delete_employees_with_age(self, age_from, age_to):
#         remove from the_back
        for idx in range(len(self.employees)-1, -1, -1): # this tricky has O(n)
            emp = self.employees[idx]
            if (age_from <= emp.age <= age_to):
                print('\nDeleting: ',emp.name)
                self.employees.pop(idx)
            else:
                print('Error Age haven`t here.')
#     search for employee by his/her name
    def find_employee_by_name(self, name):
        for emp in self.employees:
            if emp.name == name:
                return emp
            return ("Error: No employee with such a name.")
    
    def update_salary_by_name(self, name, salary):
        emp = self.find_employee_by_name(name)
        
        if (emp == None):
            print("Error: No employee with such a name.")
            print("\nPlease try again.")
        else:
            emp.salary = salary
        
```

```python
class FrontManager:
    def __init__(self):
        self.employees_manager = EmployeeManager()
#     On Start run Screen
    def print_menu(self):
        print('\nProgram Option')
        messages =[
            '1) Add a new Employee',
            '2) List of all Employee',
            '3) Delete by age range',
            '4) Update Salary by given Name Employee',
            '5) End The Program ^_^'
        ]
        print('\n'.join(messages))
        msg = F'Enter Your Choice (From 1 -> {len(messages)}):'
        return input_invalid_int(msg, 1, len(messages))
    
#     keep it run program until user end it
    def run(self):
        while True:
            choice = self.print_menu()
            
            if (choice == 1):
                self.employees_manager.add_employee()
            
            elif (choice == 2):
                self.employees_manager.list_employees()
            
            elif (choice == 3):
                age_from = input_invalid_int('Enter age From: ')
                age_to = input_invalid_int('Enter age to: ')
                self.employees_manager.delete_employees_with_age(age_from, age_to)
            
            elif (choice == 4):
                name = input('Enter Name: ')
                print(self.employees_manager.find_employee_by_name(name))
                
                salary = input_invalid_int('Enter new Salary ^_^ : ')
                self.employees_manager.update_salary_by_name(name, salary)
            
            elif (choice == 5):
                print('Goodbye see you later ^_^')
                break
```

```python
if __name__ == '__main__':
    app = FrontManager()
    app.run()
```
